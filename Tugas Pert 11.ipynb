<img src="http://www.mysqltutorial.org/wp-content/uploads/2009/12/MySQL-Sample-Database-Schema.png" />

```python
import os
import pymysql
import pandas as pd

host = 'localhost'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. Prepare a list of offices sorted by country, state, city.

```python
run('select country, state, city, officeCode, addressLine1, addressLine2, postalCode, territory from offices order by country, state, city')
```

### 2. How many employees are there in the company?

```python
run('select count(*) total_emp from employees')
```

### 3. What is the total of payments received?

```python
run('select sum(amount) as total_payment from payments')
```

### 4. List the product lines that contain 'Cars'.

```python
run("""
select productLine from productlines where productLine like '%Cars%'
""")
```

### 5. Report total payments for October 28, 2004.

```python
run("""
select sum(amount) as total_pay from payments where paymentDate = '2004-10-28'
""")
```

### 6. Report those payments greater than \$100,000.

```python
run('select * from payments where amount > 100000')
```

### 7. List the products in each product line.

```python
run('select productLine, productName from products order by productLine, productName')
```

### 8. How many products in each product line?

```python
run('select productLine, count(productName) total_product from products group by productLine')
```

### 9. What is the minimum payment received?

```python
run('select min(amount) as min_payment from payments')
```

### 10. List all payments greater than twice the average payment.

```python
run('select * from payments where amount > ( select avg(amount)*2 from payments )')
```

### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run('select avg(markup_prcnt) from (select buyPrice, MSRP, (((MSRP/buyPrice)-1)*100) as markup_prcnt from products) sub')
```

### 12. How many distinct products does ClassicModels sell?

```python
run('select count(*) as cnt_product from (select distinct productCode from products) sub')
```

### 13. Report the name and city of customers who don't have sales representatives?

```python
run('select customerName, contactFirstName, contactLastName, city from customers where salesRepEmployeeNumber is null')
```

### 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
run("""
select concat(firstName, ' ', lastName) as executive_name, jobTitle from employees
where jobTitle like '%VP%' or jobTitle like '%manager%'
""")
```
